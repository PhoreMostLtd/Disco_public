```python
import sys
import os
!{sys.executable} -m pip install numpy
!{sys.executable} -m pip install biopython
from Bio import SeqIO
import re
!{sys.executable} -m pip install pandas
import pandas as pd
from Bio.SeqUtils.IsoelectricPoint import IsoelectricPoint as IP
from Bio.SeqUtils import ProtParam


input_tbl = pd.read_csv('20200803_CPP_options.csv', sep = ",")

def get_hydropathy_score(seq):
    hydrophilicity_dict = {
        "I" :  -1.8,
        "V" :  -1.5,
        "L" :  -1.8,
        "F" :  -2.5,
        "C" :  -1.0,
        "M" :  -1.3,
        "A" :  -0.5,
        "G" : 0.0,
        "T" : -0.4,
        "S" : 0.3,
        "W" : -3.4,
        "Y" : -2.3,
        "P" : 0.0,
        "H" : -0.5,
        "E" : 3.0,
        "Q" : 0.2,
        "D" : 3.0,
        "N" : 0.2,
        "K" : 3.0,
        "R" : 3.0
    }
    Hi = []
    for AA in seq:
        Hi.append(hydrophilicity_dict.get(AA, AA))
        output = sum(Hi)
    return output

print("Name", ",", "AAseq", ",", "IEP", ",", "Charge@pH7.4", ",", "GRAVY", ",", "Length", ",", "Hydropathy_score")

for index, row in input_tbl.iterrows():
    AAseq = row['Seq']
    Name = row['Seq_name']
    X = ProtParam.ProteinAnalysis(AAseq)
    IEP = X.isoelectric_point()
    Charge_at_pH7point4 = X.charge_at_pH(7.4)
    GRAVY = X.gravy()
    Length = sum(X.count_amino_acids().values())
    Hydropathy_score = get_hydropathy_score(AAseq)
    print(Name, ",", AAseq, ",", IEP, ",", Charge_at_pH7point4, ",", GRAVY, ",", Length, ",", Hydropathy_score)
```
